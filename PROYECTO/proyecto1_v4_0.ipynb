# 3. Selección de Características

De acuerdo al artículo "A Data Feature Extraction Method Based on the NOTEARS Causal Inference Algorithm", se concluye que el uso de NOTEARS (Non-combinatorial Optimization via Trace Exponential and Augmented lagRangian for Structure learning) para seleccionar las carácteristicas con mayor causalidad en la variable objetivo optimiza el rendimiento de los modelos de regresión para predecir de forma más precisa.



## 3.1 Preparación de datos para NOTEARS

Se remueve la columna "shares" y se aplica un escalado estándar, lo cual es necesario para poder correr el algoritmo

```python
# Dataset para NOTEARS
print("\n3.1 Preparación de dataset para NOTEARS")
df_notears = df.drop(columns=["shares"])
feature_names = df_notears.columns.tolist()

# Escalado
print("\n3.2 Escalado estándar para NOTEARS")
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_notears)

X_tensor = torch.from_numpy(X_scaled).float()
```

## 3.2 Definición de funciones base para NOTEARS

```python
# Función de Aciclidad
def h_func(W):
    '''
    Función de Aciclidad
    Entrada:
        W: Matriz de pesos
    Salida:
        Valor de la función de pérdida
    '''
    d = W.shape[0]
    return torch.trace(torch.matrix_exp(W * W)) - d

# Función de Pérdida
def loss_func(X, W):
    '''
    Función de Pérdida
    Entrada:
        X: Matriz de datos
        W: Matriz de pesos
    Salida:
        Valor de la función de pérdida
    '''
    n = X.shape[0]
    return 0.5 / n * torch.sum((X - X @ W) ** 2)

# Función de implementación de Notears
# Optimización continua, Penalización cuadrática y Multiplicador de Lagrange actualizado iterativamente
def notears_augmented(X, lambda1=0.05, max_iter=100, lr=0.01, rho_init=1.0, rho_max=1e6, h_tol=1e-8):
    '''
    Función de implementación de Notears
    Entrada:
        X: Matriz de datos
        lambda1: Parámetro de regularización
        max_iter: Número máximo de iteraciones
        lr: Tasa de aprendizaje
        rho_init: Valor inicial del multiplicador de Lagrange
        rho_max: Valor máximo del multiplicador de Lagrange
        h_tol: Tolerancia para la función de pérdida
    Salida:
        Matriz de pesos
    '''
    n, d = X.shape
    W = torch.zeros(d, d, requires_grad=True)

    rho = rho_init
    alpha = 0.0

    optimizer = optim.Adam([W], lr=lr)

    for it in range(max_iter):
        for _ in range(100):  # inner optimization
            optimizer.zero_grad()

            loss = loss_func(X, W)
            h = h_func(W)
            l1 = lambda1 * torch.sum(torch.abs(W))

            obj = loss + l1 + alpha * h + 0.5 * rho * h * h
            obj.backward()
            optimizer.step()

        with torch.no_grad():
            h_val = h_func(W).item()
            print(f"Iter {it:03d} | Loss: {loss.item():.4f} | h(W): {h_val:.6e}")

            alpha += rho * h_val

            if abs(h_val) <= h_tol:
                break

            rho = min(rho * 10, rho_max)

    return W.detach().numpy()
```

## 3.3 Ejecución de algoritmo NOTEARS

Se ejecuta el algoritmo NOTEARS para obtener la matriz de pesos

```python
X_sub = X_tensor[:20000]  # Valor conservador para acelerar el cómputo

W_est_mdpi = notears_augmented(
    X_sub,
    lambda1=0.01,
    max_iter=20
)

# Para eliminar ruido, se estable umbral
#W_est_mdpi[np.abs(W_est_mdpi) < 0.1] = 0
```

## 3.4 Extracción de pesos causales

Se extraen los pesos de cada variable con la variable objetivo "log_shares"

```python
target = "log_shares"
target_idx = feature_names.index(target)

causal_weights = []

for i, feature in enumerate(feature_names):
    if feature == target:
        continue

    weight = W_est_mdpi[i, target_idx]
    causal_weights.append((feature, weight))
```

## 3.5 Ordenamiento del dataframe por causalidad

Se ordena el dataframe por causalidad y se agrega la columna "Abs_Weight" para obtener los valores absolutos de los pesos calculados

```python
causal_df = (
    pd.DataFrame(causal_weights, columns=["Feature", "Causal_Weight"])
    .assign(Abs_Weight=lambda x: x["Causal_Weight"].abs())
    .sort_values("Abs_Weight", ascending=False)
    .reset_index(drop=True)
)
#print("\n3.5 Ordenamiento del dataframe por causalidad")
#causal_df = (
#    pd.DataFrame(causal_weights, columns=["Feature", "Causal_Weight"])
#    .assign(Abs_Weight=lambda x: x["Causal_Weight"].abs())
#    .query("Abs_Weight > 0.05") # Solo variables con peso real
#    .sort_values("Abs_Weight", ascending=False)
#)

#causal_df.head(15)
causal_df
```

# 4. Evaluación de Modelos de Regresión

## 4.1 Función para evaluación de modelos

```python
def evaluate_model(name, model, X_test, y_test):
    '''
    Funcion para evaluar modelos de regresión
    Entrada:
        name: nombre del modelo
        model: modelo de regresión
        X_test: conjunto de datos de prueba
        y_test: variable objetivo de prueba
    Salida:
        Diccionario con métricas de evaluación
    '''
    y_pred = model.predict(X_test)

    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)

    return {
        "Modelo": name,
        "R2": r2_score(y_test, y_pred),
        "MAE": mean_absolute_error(y_test, y_pred),
        "MSE": mse,
        "RMSE": rmse
    }
```

### 4.2.1 Definición de Pipelines

Se define el pipeline para los modelos:

*   SVM con StandarScaler (centro en 0. y varianza 1) para evitar dominio de variables grandes y que el hiperplano se deforme.
*   AdaBoost con MinMaxScaler para estabilizar valores con escalas muy distintas, con 200 árboles y tasa de aprendizaje de 5%.



```python
models_no_causal = {
    "RandomForest": Pipeline([
        ("model", RandomForestRegressor(
            n_estimators=300,
            max_depth=15,
            random_state=42,
            n_jobs=-1
        ))
    ]),

    "SVM": Pipeline([
        ("scaler", StandardScaler()),
        ("model", SVR(
            kernel="rbf",
            C=10,
            epsilon=0.1
        ))
    ]),

    "AdaBoost": Pipeline([
        ("scaler", MinMaxScaler()),
        ("model", AdaBoostRegressor(
            n_estimators=200,
            learning_rate=0.05,
            random_state=42
        ))
    ])
}
```

### 4.3.3 Pipelines con features causales

Se define el pipeline para los modelos:

*   Random Forest con 300 árboles de profundidad 15 utilizando todos los core disponibles.
*   SVM con StandarScaler (centro en 0. y varianza 1) para evitar dominio de variables grandes y que el hiperplano se deforme.
*   AdaBoost con MinMaxScaler para estabilizar valores con escalas muy distintas, con 200 árboles y tasa de aprendizaje de 5%.



```python
print("\n4.3.3 Definición de Pipelines para modelos de regresión:")
models_causal = {
    "RandomForest": Pipeline([
        ("model", RandomForestRegressor(
            n_estimators=300,
            max_depth=15,
            random_state=42,
            n_jobs=-1
        ))
    ]),

    "SVM": Pipeline([
        ("scaler", StandardScaler()),
        ("model", SVR(
            kernel="rbf",
            C=10,
            epsilon=0.1
        ))
    ]),

    "AdaBoost": Pipeline([
        ("scaler", MinMaxScaler()),
        ("model", AdaBoostRegressor(
            n_estimators=300,
            learning_rate=0.05,
            random_state=42
        ))
    ])
}
```

# 6. Análisis de resultados

Análisis de resultados y comparativo de métricas:

## 6.1 Impacto positivo: Random Forest y SVM
Para estos dos modelos, la selección de características basada en causalidad (NOTEARS) fue exitosa:
* Random Forest: Es el modelo más optimo. Su $R^2$ subió de 0.134 a 0.147 y su RMSE bajó de 0.861 a 0.854. Esto se explica ya que al reducir el ruido y eliminar correlaciones espurias, el bosque de árboles pudo encontrar reglas de decisión más generales y menos ajustadas a "coincidencias" estadísticas del dataset original.
* SVM (Support Vector Machine): Tuvo la mejora relativa más alta en $R^2$, subiendo de 0.079 a 0.099 (casi el 25% en su capacidad explicativa). Esto se explica ya que SVM es muy sensible a la dimensionalidad por lo que al darle sólo las variables que NOTEARS identificó como causales, el hiperplano de regresión se vuelve mucho más estable y preciso.

## 6.2. Impacto Negativo: AdaBoost
AdaBoost tiene una degradación severa, pasando de un $R^2$ positivo (0.021) a uno negativo (-0.085). Un $R^2$ negativo significa que el modelo es peor que simplemente predecir el promedio de los datos.¿Por qué falló?:
1.  Dependencia de Proxies: AdaBoost construye modelos de forma secuencial corrigiendo errores. Es posible que este algoritmo estuviera utilizando variables que NO son causales pero que actúan como "proxies" (representantes) de información compleja. Al eliminarlas con NOTEARS, AdaBoost perdió las muletas estadísticas en las que se apoyaba.
2.  Sensibilidad al Weak Learner: Si el estimador base de AdaBoost (el árbol pequeño) no tiene suficientes variables para encontrar cortes significativos, el proceso de "boosting" no puede arrancar, resultando en un modelo que no aprende nada útil.

## 6.3 Interpretación de Métricas

* MAE (Error Absoluto Medio): Observamos que para el modelo NOTEARS de SVM y RF, el MAE ronda los 0.62 - 0.63. Dado que estamos prediciendo el log(shares), un error de 0.6 significa que nuestras predicciones suelen estar en el mismo orden de magnitud que el valor real, lo cual es un logro considerable para un fenómeno tan errático como la viralidad.

* RMSE vs MSE: El RMSE sigue siendo alto (~0.85), lo que indica que todavía tenemos errores grandes en las noticias extremadamente virales (outliers). Esto confirma que, aunque las variables causales ayudan, la viralidad extrema sigue teniendo un componente aleatorio o externo no capturado en el dataset.

## 6.4 Conclusiones

1. **Eficiencia**: El modelo Random Forest + NOTEARS es ahora más sencillo (menos variables) y más preciso que el modelo original.

2. **Robustez Causal**: Se verifica que variables como **is_weekend** o **kw_avg_avg** no solo están correlacionadas, sino que el algoritmo NOTEARS les asigna un peso estructural.

3. **Lección Aprendida**: La caída de AdaBoost indica que la selección de características no es "talla única". Los modelos que dependen fuertemente de la combinación de muchas variables débiles pueden sufrir si se reduce demasiado el espacio de búsqueda.
